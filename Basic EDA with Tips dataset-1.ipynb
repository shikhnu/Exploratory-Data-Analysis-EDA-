### Scaling

```python
df.head()
```

```python
#scaling with Zscore method: 
df['z_score'] = (df['total_bill']-df['total_bill'].mean())/df['total_bill'].std()
```

```python
df.head()
```

```python
f,ax=plt.subplots(1,2,figsize=(15,5))
sns.distplot(df['total_bill'], ax=ax[0])
sns.distplot(df['z_score'], ax=ax[1])
plt.show()
```

```python
print('Skewness of total bill: ', df['total_bill'].skew())
print('Skewness of scaled total bill(z_score): ', df['z_score'].skew())
print('kurtosis of total bill: ', df['total_bill'].kurt())
print('kurtosis of scaled total bill(z_score): ', df['z_score'].kurt())
```

#### min-max scaling / normalization
- bringing the values between 0 to 1

```python
df['min_max'] = (df['total_bill']-df['total_bill'].min())/(df['total_bill'].max()-df['total_bill'].min())
df.head()
```

```python
f,ax=plt.subplots(1,3,figsize=(15,5))
sns.distplot(df['total_bill'], ax=ax[0])
sns.distplot(df['z_score'], ax=ax[1])
sns.distplot(df['min_max'], ax=ax[2])
plt.show()
```

```python
print('Skewness of total bill: ', df['total_bill'].skew())
print('Skewness of scaled total bill(z_score): ', df['z_score'].skew())
print('Skewness of scaled total bill(min_max): ', df['min_max'].skew())
print('kurtosis of total bill: ', df['total_bill'].kurt())
print('kurtosis of scaled total bill(z_score): ', df['z_score'].kurt())
print('kurtosis of scaled total bill(min_max): ', df['min_max'].kurt())
```
